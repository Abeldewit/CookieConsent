```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import ast

df = pd.read_csv('assets/DataFrame.csv', converters={"Host": ast.literal_eval,
                                                     "cookie_first": ast.literal_eval,
                                                     "cookie_second": ast.literal_eval
                                                     })
```

```python
counter_1 = 0
counter_2 = 0
counter_3 = 0

for index, row in df.iterrows():
    if (len(row['cookie_first']) > 0) & (len(row['cookie_second']) > 0):
        counter_3 += 1
    if len(row['cookie_first']) > 0:
        counter_1 += 1
    if len(row['cookie_second']) > 0:
        counter_2 += 1

    
print("Found {} intial cookies, {} when accepting cookies, and {} with both".format(counter_1, counter_2, counter_3))
```

```python
providers = ['squarespace', 'wix', 'weebly', 'jimdo', 'shopify',
             'bigcommerce', 'webnode', 'site123', 'wordpress',
             'medium-com', 'webflow', 'joomla']

site_stats = {}
for p in providers:
    mask = df['Host'].apply(lambda x: p in x)
    p_df = df[mask]
    site_stats[p] = len(p_df)

fig, ax = plt.subplots()
plt.bar(range(len(site_stats)), list(site_stats.values()), align='center')
plt.xticks(range(len(site_stats)), list(site_stats.keys()), rotation='vertical')
plt.show()
```

```python
import math
for p in site_stats:
    print(p, math.floor((12-len(p))/6)*'\t'+'\t', site_stats[p])
```

```python
import base64
import binascii

count = 0
for index, row in df.iterrows():
    if len(row['cookie_first']) != 0:
        cookie = row['cookie_first'].pop()
        for key, value in cookie.items():
            if key == 'value':
                try:
                    base64_bytes = value.encode('ascii')
                    message_bytes = base64.b64decode(base64_bytes)
                    message = message_bytes.decode('ascii')
                    print(key, '->', message)
                except binascii.Error:
                    print(key, '->', value, '*')
                except UnicodeDecodeError:
                    print(key, '->', value, '*')
            else:
                # print(key, '->', value)
                print('', end='')
        count+=1
    if count > 100:
        break
```

```python
counts = {'gid': 0, 'fbp': 0, 'consent': 0, 'gtag': 0, 'cfduid': 0}


for index, row in df.iterrows():
    cookie = row['cookie_first']
    if len(cookie) > 0:
        cookie = cookie[0]
        print(row['Website'])
        for k, v in cookie.items():
            found = False
            if k == 'name':
                for _id in counts.keys():
                    if _id in v.lower():
                        counts[_id] += 1
                        found = True
                        continue

                if not found:
                    print(cookie['name'], ':', cookie['value'])              

        print('-'*20)

print(counts)
```

```python
unique_names = []

for cookie in df['cookie_first']:
    if len(cookie) > 0:
        cookie = cookie[0]
        for k, v in cookie.items():
            if k == 'name':
                if v.lower() not in unique_names:
                    unique_names.append(v.lower())

for name in unique_names:
    if name[-2:] == 'id':
        print(name)

print('-*-'*10)
for name in unique_names:
    if 'geo' in name:
        print(name)
        
print('-*-'*10)
for name in unique_names:
    if 'cookie' in name:
        print(name)

print('-*-'*10)
for name in unique_names:
    if 'gdpr' in name:
        print(name)
```
